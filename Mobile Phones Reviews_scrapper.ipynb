```python
import time
import selenium
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver as wb
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
webD = wb.Chrome(r'C:\Users\saurav\Desktop\chromedriver.exe',chrome_options=chrome_options)
webD.get('https://www.amazon.in/')


#go to mobiles by clicking on Mobile using its xpath
start = time.process_time()
webD.find_element_by_xpath('//*[@id="nav-xshop"]/a[1]').click()



#store the links of mobile brands in list name Mobile_brands
Mobile_brands=[]
first_half=webD.find_element_by_xpath('/html/body/div[2]/div[4]/div/div[1]/div/div[1]/div[14]/div/div/div[5]')
second_half=webD.find_element_by_xpath('/html/body/div[2]/div[4]/div/div[1]/div/div[1]/div[14]/div/div/div[6]')
first=first_half.find_elements_by_class_name('bxc-grid__column.bxc-grid__column--3-of-12.bxc-grid__column--light')
second=second_half.find_elements_by_class_name('bxc-grid__column.bxc-grid__column--2-of-12.bxc-grid__column--light')
for el in first:
    Mobile_brands.append(el.find_element_by_tag_name('a').get_property('href'))

Mobile_brands.append(second[0].find_element_by_tag_name('a').get_property('href'))



#store the links of 100 mobiles, from different brands in the list named List_of_100_mobiles_from_different_brands
List_of_100_mobiles_from_different_brands=[]
for element in Mobile_brands:
    webD.get(element)
    time.sleep(2)
    page=webD.find_element_by_class_name('s-result-list.s-col-3.s-result-list-hgrid.s-height-equalized.s-grid-view.s-text-condensed')
    for i in range(23):
        try:
            List_of_100_mobiles_from_different_brands.append(page.find_element_by_id('result_'+str(i)).find_element_by_tag_name('a').get_property('href'))
        except:
            continue

            
            
# going to the page of every mobile phones through there links(List_of_100_mobiles_from_different_brands)
# And appending mobile names and reviews of mobiles in list a and b 
a=[]
b=[]
for el in List_of_100_mobiles_from_different_brands:
    webD.get(el)
    time.sleep(2)
    try:
        b.append(webD.find_element_by_id('cm-cr-dp-review-list').find_element_by_class_name('a-row.a-spacing-small.review-data').text)
        a.append(webD.find_element_by_tag_name('h1').text)
    except:
        continue


        
#buinding the data frame 
print('\n\ntime taken in scrapping data:', time.process_time() - start)
data = {'Mobile Name':a, 'reviews':b}
dataset = pd.DataFrame(data)




#generating the Excel file from data frame
dataset.to_excel(r'C:\Users\saurav\Desktop/Mobile_Reviews.xlsx',index = False)
dataset.head()
```
